```python
#Euclidean
def homework(train_X, train_y, test_X):
    k = 3
    pred_y = []
    for i in range(test_X.shape[0]):  #need to calculate the distances between each single vector in text_X and all those in train_X
        d = np.linalg.norm(test_X[i] - train_X, axis = 1) # calculate all the distances 
        l = train_y[d.argsort()][:k] #the k labels that have the shortest distances
        counts = np.bincount(l) #count the number of appearance of each label
        pred_y = np.append(pred_y, counts.argmax()) # check the majority and append that label into the predicted result
    return pred_y
```

```python
#Manhattan
def homework(train_X, train_y, test_X):
    # WRITE ME   
    k = 3
    pred_y = []
    for i in range(test_X.shape[0]): 
        d = np.add.reduce(abs(test_X[i] - train_X), axis = 1)
        l = train_y[d.argsort()][:k] 
        counts = np.bincount(l)
        pred_y = np.append(pred_y, counts.argmax())
    return pred_y
```

```python
validate_homework() #Manhattan
```

```python
#Cosine
def homework(train_X, train_y, test_X):
    # WRITE ME   
    k = 3
    pred_y = []
    for i in range(test_X.shape[0]):
        
        na = test_X[i] / np.linalg.norm(test_X[i])
        nb = train_X / np.linalg.norm(train_X, axis = 1).reshape(-1,1)
        d = np.dot(na, nb.T)
        l = train_y[d.argsort()][-k:]

        counts = np.bincount(l)
        pred_y = np.append(pred_y, counts.argmax())
    
    return pred_y
```

### PCA method implementation

```python
#PCA method using S, V and D
def svd_pca(data, k):
    """Reduce DATA using its K principal components."""
    data = data.astype("float64")
    data -= np.mean(data, axis=0)
    U, S, V = np.linalg.svd(data, full_matrices=False)
    return U[:,:k].dot(np.diag(S)[:k,:k])
```
